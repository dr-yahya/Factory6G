UMi (Urban Micro) is ideal for smart factories because:

Designed for small cell deployments with low base station heights (typically 10m)
Covers short distances (up to 5km), perfect for indoor/factory environments
Models dense, cluttered environments with many reflections - typical of factory floors with machinery, equipment, and metallic structures
Supports both line-of-sight (LOS) and non-line-of-sight (NLOS) scenarios common in industrial settings
The "o2i_model='low'" parameter is appropriate for factory buildings with typical wall penetration losses
