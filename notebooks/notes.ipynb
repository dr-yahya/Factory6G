The **6.425-7.125 GHz band** represents the most significant advancement for 6G industrial applications, providing 700 MHz of harmonized spectrum across ITU regions. Path loss characteristics follow the Close-In (CI) model: PL(dB) = 32.4 + 10n log₁₀(d) + Xσ, where path loss exponent n = 1.4-1.7 for LOS scenarios and n = 3.2-3.4 for NLOS conditions. Shadow fading ranges from 2.7-2.9 dB (LOS) to 8.6-10.1 dB (NLOS).

**Mid Band 6-24GHz (Centimeter Wave)**

The **7-24 GHz range** (also designated as FR3) provides the optimal balance between coverage and capacity for smart factory applications. These frequencies utilize the Alpha-Beta-Gamma (ABG) path loss model: PL(dB) = 10α log₁₀(d) + β + 10γ log₁₀(fc) + Xσ, with distance dependence α = 1.7 (LOS) and 3.2 (NLOS), and frequency dependence γ = 1.9 (LOS) and 3.4 (NLOS).

This band supports high-capacity factory automation, precision manufacturing control, and real-time industrial sensing with superior penetration characteristics compared to mmWave frequencies. The extended mid-band offers up to 20 GHz of available spectrum for ultra-high capacity applications.

**High Band mmWave and Sub-THz (24GHz+)**

Sub-THz bands (100-300 GHz) represent the frontier of 6G technology, with ITU-R identifying specific allocations including 102-109.5 GHz, 151.5-164 GHz, 209-226 GHz, and 252-275 GHz. These frequencies enable revolutionary applications like holographic communication and ultra-precision manufacturing control, supporting data rates up to 1 Tbps with extremely low latency (microseconds), though limited to very short ranges (\<100 meters).

## Channel models for indoor factory environments

**3GPP Standardized Models**

The **3GPP TR 38.901 Indoor Factory (InF) model** serves as the foundation for industrial channel modeling, supporting frequencies from 0.5 to 100 GHz with bandwidth up to 1 GHz. This geometry-based stochastic model implements clustered multipath propagation with both LOS and NLOS scenarios, accommodating distances from 1-150 meters typical in factory environments.

For sub-6GHz applications, the Close-In (CI) model provides: PL(dB) = 32.4 + 10n log₁₀(d) + Xσ, with path loss exponents n = 1.4-1.7 (LOS) and n = 3.2-3.4 (NLOS). Shadow fading standard deviation ranges from 2.7-2.9 dB (LOS) to 8.6-10.1 dB (NLOS).

**Factory-Specific Propagation Characteristics**

Industrial environments present unique propagation challenges due to metallic equipment, machinery, and structural elements. **Metal surfaces create enhanced reflections with reflection coefficients approaching -1**, while large equipment causes significant shadowing (10-30 dB additional loss). Corridors and aisles between equipment create waveguide effects that can enhance or degrade signal propagation.

Key propagation parameters include:
- **RMS Delay Spread**: 50-200 ns (sub-6GHz), 10-50 ns (mmWave), 5-20 ns (THz)
- **Angular Spread**: 10-45° (azimuth), 5-20° (elevation)
- **K-Factor**: 10-20 dB (LOS), 0-5 dB (NLOS)
- **Path Loss Exponent**: 1.4-1.7 (LOS), 3.2-3.4 (NLOS) for sub-6GHz

**THz and mmWave Characteristics**

At THz frequencies, propagation exhibits significant sparsity with fewer dominant multipath components. **Diffuse scattering becomes dominant due to surface roughness effects**, while atmospheric absorption and molecular resonances create frequency-dependent losses. Path loss exponents range from ~2.0 (LOS) to 4.6 (NLOS), with delay spreads compressed to 4.27 ns due to the sparse nature of THz propagation.

## Physical layer parameters for 6G

**Subcarrier Spacing and FFT Configuration**

6G extends 5G NR's subcarrier spacing from the current 15-240 kHz range to **15-960 kHz, potentially reaching 3840 kHz for sub-THz applications**. Higher subcarrier spacing enables shorter symbol durations, critical for ultra-low latency industrial IoT applications. Industrial applications will benefit from adaptive selection between narrow spacing (15-60 kHz) for coverage and wide spacing (480-960 kHz) for ultra-low latency.

**FFT sizes scale from 5G's maximum 4096 points to 8192 and potentially 16384 points**, supporting wider bandwidths and higher-order spatial multiplexing. The relationship FFT size = sampling frequency / subcarrier spacing enables flexible bandwidth allocation for diverse industrial applications.

**Modulation and MIMO Evolution**

6G extends modulation schemes from 5G's maximum 256QAM to **1024QAM and potentially higher orders (2048QAM, 4096QAM)** for high-SNR industrial scenarios. Adaptive modulation balances throughput and reliability based on channel conditions and application requirements.

**MIMO configurations evolve from current 64×64 deployments to 128×128 and 256×256 massive MIMO**, with research demonstrating systems up to 4096 antenna elements. XL-MIMO (Extremely Large MIMO) and Giga-MIMO represent the next evolution, enabling unprecedented spatial multiplexing and beamforming precision for industrial applications.

**Waveform Candidates**

Beyond OFDM, 6G evaluates alternative waveforms including:
- **FBMC (Filter Bank Multi-Carrier)**: Superior spectral containment
- **GFDM (Generalized Frequency Division Multiplexing)**: Highest spectral efficiency
- **UFMC (Universal Filtered Multi-Carrier)**: Fragmented spectrum optimization
- **F-OFDM (Filtered OFDM)**: Enhanced spectral efficiency with OFDM compatibility
