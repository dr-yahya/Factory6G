# 5G OFDM System Configuration and Simulation Results

## System Parameters

### Radio Frequency Configuration
- **Carrier Frequency**: 3.5 GHz
- **FFT Size**: 128
- **Subcarrier Spacing**: 30 kHz

### OFDM Frame Structure
- **Number of OFDM Symbols**: 14
- **Cyclic Prefix Length**: 20
- **Pilot OFDM Symbol Indices**: [2, 11]

### Antenna Configuration
- **Number of Base Station Antennas**: 8
- **Number of User Terminals**: 4
- **Number of Antennas per User Terminal**: 1

### Modulation and Coding
- **Bits per Symbol**: 2
- **Code Rate**: 0.5
